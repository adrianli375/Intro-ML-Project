# Demo 2: A Machine Learning Regression Task

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.compose import make_column_transformer
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_validate, train_test_split, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.svm import SVR
```

## Dataset Description

This [dataset](https://www.kaggle.com/datasets/mirichoi0218/insurance) is from Kaggle. It is to predict the medical cost of patients given their own attributes. Let's read the dataset and split it into train and test portions: 

```python
df = pd.read_csv("../datasets/insurance.csv")

train_df, test_df = train_test_split(df, test_size=0.3, random_state=7413)
```

Let's have a glimpse at the training set: 

```python
train_df
```

## Split into X and y

```python
X_train, y_train = train_df.drop(columns=["charges"]), train_df["charges"]
X_test, y_test = test_df.drop(columns=["charges"]), test_df["charges"]
```

## Preprocessing

Scale the features `age`, `bmi`. One-hot encode the binary features `sex`, `male`. One-hot encode the categorical feature `region`. 

```python
numeric_feats = ["age", "bmi"]
binary_feats = ["sex", "smoker"]
categorical_feats = ["region"]
pass_feats = ["children"]
```

```python
preprocessor = make_column_transformer(
    (StandardScaler(), numeric_feats), 
    (OneHotEncoder(drop="if_binary"), binary_feats), 
    (OneHotEncoder(drop="first"), categorical_feats), 
    ("passthrough", pass_feats)
)
```

```python
X_train_transformed = preprocessor.fit_transform(X_train)
```

Let's have a look at the transformed training dataset (`X_train_transformed`): 

```python
pd.DataFrame(X_train_transformed, columns=preprocessor.get_feature_names_out())
```

## Baseline model

We will use `LinearRegression` as our baseline model for reference. 

```python
baseline_model = make_pipeline(
    preprocessor, 
    LinearRegression()
)
```

```python
baseline_model.fit(X_train, y_train)
```

Let's evaluate our model using the $R^2$-score. 

```python
baseline_score = baseline_model.score(X_train, y_train)
baseline_score
```

We got an $R^2$-score of `0.763`. Can we perform better?

## Modelling

There are various regression models we can adopt. In this demo, I will be adopting the following models: 
1.  Ridge regression (L2-regularized loss function)
2.  Lasso regression (L1-regularized loss function)
3.  Random Forest Regression
4.  Gradient Boosting Regressor

We will be using the cross-validation $R^2$-scores for our evaluation. 

```python
def cv_val_score(model):
    return cross_validate(model, X_train, y_train, cv=10, return_train_score=True)["test_score"].mean()
```

### Ridge regression

```python
model_ridge = make_pipeline(
    preprocessor, 
    Ridge(random_state=16)
)
```

```python
ridge_score = cv_val_score(model_ridge)
ridge_score
```

### Lasso regression

```python
model_lasso = make_pipeline(
    preprocessor, 
    Lasso(random_state=214)
)
```

```python
lasso_score = cv_val_score(model_lasso)
lasso_score
```

### Random Forest regression

```python
model_rf = make_pipeline(
    preprocessor, 
    RandomForestRegressor(random_state=215)
)
```

```python
rf_score = cv_val_score(model_rf)
rf_score
```

### Gradient Boosting regression

```python
model_gb = make_pipeline(
    preprocessor, 
    GradientBoostingRegressor(random_state=216)
)
```

```python
gb_score = cv_val_score(model_gb)
gb_score
```

### Summary of the regression models tested

Let's plot the performance of our different regression models in a bar plot: 

```python
models = ["LinearRegression", "Ridge", "Lasso", "RandomForestRegressor", "GradientBoostingRegressor"]
model_scores = [baseline_score, ridge_score, lasso_score, rf_score, gb_score]
colours = ["red", "orange", "yellow", "green", "blue"]

plt.xlabel("Models")
plt.ylabel("$R^2$ score")
plt.title("Performance of different regression models")
plt.xticks(rotation=30)

plt.bar(models, model_scores, color=colours)

plt.show()
```

It seems that `GradientBoostingRegressor` is the best performing model on the training set with a validation $R^2$-score of `0.866`. 

## Hyperparameter optimization

Now, we tune the hyperparameters in the model and let's see if we can achieve an even better performance!

```python
param_grid = {
    "gradientboostingregressor__loss": ["squared_error", "huber"], 
    "gradientboostingregressor__learning_rate": [1e-3, 1e-2, 1e-1], 
    "gradientboostingregressor__n_estimators": [32, 64, 128]
}
```

```python
grid_search = GridSearchCV(model_gb, param_grid, n_jobs=-1, return_train_score=True)
grid_search.fit(X_train, y_train)
```

```python
grid_search_cv_results = pd.DataFrame(grid_search.cv_results_)
grid_search_cv_results
```

Let's see which model performs the best!

```python
best_params = grid_search.best_params_
best_params
```

Let's find out the validation score of this best performing model: 

```python
model_index = grid_search_cv_results[
    grid_search_cv_results["params"] == best_params
].index.tolist()[0]
grid_search_cv_results.at[model_index, "mean_test_score"]
```

This model performed slightly better: we obtained a validation $R^2$-score of `0.8716`. 

## Evaluate model test score

```python
y_pred = grid_search.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

Our final model obtained a test score of `0.8289`.
