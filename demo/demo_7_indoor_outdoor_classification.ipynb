## Imports
<hr>

```python
from datetime import datetime
from heic2png import HEIC2PNG
import math
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
import random
import re
import shutil
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import image_dataset_from_directory, plot_model
from tqdm import tqdm

random.seed(701)
```

## Global Variables
<hr>

```python
IMG_SIZE = (256, 256)
INPUT_SHAPE = (256, 256, 3)
N_IMAGE_PER_CLASS = 60
VALIDATION_SIZE = 0.1

MODEL_PATH = '../models/demo_7_img_classifier.h5'
```

## Utils
<hr>

```python
def visualize_image(file_path: str):
    img = Image.open(file_path).convert('RGB')
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
def convert_img_to_png(img_file_path: str):
    if re.search(r'.heic$', img_file_path, flags=re.IGNORECASE):
        img = HEIC2PNG(img_file_path)
        img.save()
        os.remove(img_file_path)
    elif re.search(r'.jp(e)?g$', img_file_path, flags=re.IGNORECASE):
        new_path = re.sub(r'.jp(e)?g', '.png', img_file_path, flags=re.IGNORECASE)
        img = Image.open(img_file_path)
        img.save(new_path)
        os.remove(img_file_path)
```

```python
def compress_img(file_path: str):
    img = Image.open(file_path)
    new_img = img.resize(IMG_SIZE)
    new_file_path = re.sub(r'raw', 'processed', file_path, flags=re.IGNORECASE)
    new_img.save(new_file_path)
```

```python
def generate_random_file_name():
    return f'{hex(random.randint(0, 16 ** 6 - 1))}.png'
```

```python
def predict_image_class(img_path: str, model, show_img=True):
    img = image.load_img(img_path, target_size=IMG_SIZE)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0
    predictions = model.predict(img_array)
    predicted_class = 'outdoor' if np.argmax(predictions) else 'indoor'
    print(f'The predicted class of the image is {predicted_class}')
    if show_img:
        visualize_image(img_path)
```

### Step 3 – Resizing images into the same size

```python
new_dirs = ['../img/demo_7/processed', '../img/demo_7/processed/indoor', '../img/demo_7/processed/outdoor']
for dir in new_dirs:
    if not os.path.exists(dir):
        os.mkdir(dir)
```

```python
imgs_path = ['../img/demo_7/raw/indoor', '../img/demo_7/raw/outdoor']
for path in imgs_path:
    files = os.listdir(path)
    for i in tqdm(range(len(files))):
        file = files[i]
        if not re.search(r'.png$', file):
            continue
        compress_img(os.path.join(path, file))
shutil.rmtree('../img/demo_7/raw')
```

### Step 4 – Augmenting new images

In this step, we augment new images (using existing images) through rotation and/or flipping. 

```python
imgs_path = ['../img/demo_7/processed/indoor', '../img/demo_7/processed/outdoor']
for path in imgs_path:
    files = os.listdir(path)
    n_files = len(files)
    for i in tqdm(range(max(N_IMAGE_PER_CLASS - n_files + 1, 0))):
        file = random.choice(files)
        if not re.search(r'.png$', file):
            continue
        with Image.open(os.path.join(path, file)) as im:
            augmented_img = im.transpose(method=0)
            augmented_img.save(os.path.join(path, generate_random_file_name()))
```

### Step 5 – Splitting into training and validation sets

In this step, we split the image data into training and validation sets. This is to prevent the trained model from overfitting. 

```python
new_dirs = ['../img/demo_7/train', '../img/demo_7/train/indoor', '../img/demo_7/train/outdoor', 
            '../img/demo_7/valid', '../img/demo_7/valid/indoor', '../img/demo_7/valid/outdoor']
for dir in new_dirs:
    if not os.path.exists(dir):
        os.mkdir(dir)
```

```python
imgs_path = ['../img/demo_7/processed/indoor', '../img/demo_7/processed/outdoor']
for path in imgs_path:
    files = os.listdir(path)
    n_files = len(files)
    valid_size = math.floor(n_files * VALIDATION_SIZE)
    valid_sample = random.sample(list(range(n_files)), k=valid_size)
    for i in tqdm(range(n_files)):
        file = files[i]
        if not re.search(r'.png$', file):
            continue
        class_ = 'indoor' if re.search(r'indoor$', path) else 'outdoor'
        with Image.open(os.path.join(path, file)) as im:
            if i not in valid_sample:
                file_path = os.path.join(path.replace('processed', 'train'), f'train_{i:03d}_{class_}.png')
            else:
                file_path = os.path.join(path.replace('processed', 'valid'), f'valid_{i:03d}_{class_}.png')
            im.save(file_path)
shutil.rmtree('../img/demo_7/processed')
```

## Modelling
<hr>

Now, let's create our convolutional neural network (CNN) for classifying indoor and outdoor images. Let's create a model from the `tensorflow.keras` library and build the model with convolutional layers and max pooling layers. 

```python
train_dataset = image_dataset_from_directory('../img/demo_7/train', image_size=IMG_SIZE)
valid_dataset = image_dataset_from_directory('../img/demo_7/valid', image_size=IMG_SIZE)
```

<hr>
Next, we build the model and obtain the model summary. 

```python
model = Sequential(
    [
        Conv2D(16, (3, 3), activation='relu', input_shape=INPUT_SHAPE), 
        MaxPooling2D((2, 2)), 
        Conv2D(16, (3, 3), activation='relu'), 
        MaxPooling2D((2, 2)), 
        Conv2D(16, (3, 3), activation='relu'), 
        MaxPooling2D((2, 2)), 
        Conv2D(32, (3, 3), activation='relu'), 
        MaxPooling2D((2, 2)), 
        Conv2D(32, (3, 3), activation='relu'), 
        MaxPooling2D((2, 2)), 
        Conv2D(64, (3, 3), activation='relu'), 
        MaxPooling2D((2, 2)), 
        Flatten(), 
        Dense(4, activation='relu'), 
        Dropout(0.2), 
        Dense(2, activation='softmax')
    ]
)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

<hr>
Then, it's time to train our model! It make take some time if the model is first trained...

```python
if not os.path.exists(MODEL_PATH):
    model.fit(train_dataset, batch_size=32, epochs=16)
else:
    print(datetime.now(), 'Model has already been trained!')
```

<hr>
Finally, let's save our model for predictions on unseen images. 

```python
model.save(MODEL_PATH)
```

## Model Evaluation
<hr>

Let's try to evaluate our model performance using the validation set and see how well it performs!

```python
if os.path.exists(MODEL_PATH):
    print(datetime.now(), 'Loading model...')
    model = load_model(MODEL_PATH)
valid_loss, valid_accuracy = model.evaluate(valid_dataset)
print(f'The validation accuracy is {valid_accuracy:.2f}')
```

## Conclusion
<hr>

The current image classification model's accuracy is low, at only 51%, which is no different from purely random guessing. Despite adding more convolutional layers to the architecture,the model might not be effectively capturing the distinguishing features of the indoor and outdoor venues in the dataset.

The limitations of the model can be attributed to several factors. First and foremost, the dataset size might not be sufficient to train a complex model effectively. With only a limited number of images (specifically, only 60 images for each class in the training and validation datasets), the model may struggle to generalize well to unseen examples due to underfitting the training data. Furthermore, the complexity of the model architecture might also contribute to the limited performance. Having a large number of parameters could lead to overfitting, especially when the dataset is small. 

In summary, understanding the limitations of the current image classification model is an essential step towards devising strategies for improvement. By tackling the challenges related to data quality, model complexity, and optimization, we can work towards building a more accurate and reliable model for classifying indoor and outdoor venues. Continuous experimentation and refinement will be key to achieving the desired performance levels in the image classification task.
